```python
import cv2
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('./../../data/ObjectsDetected/2023-06-13/10998877/entry/cell phone.png')
img2 = cv2.imread('./../../data/ObjectsDetected/2023-06-13/10998877/exit/cell phone.png')

#write image correctly
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Display the images in the subplots
fig.suptitle('ORIGINAL IMAGE VERSIONS')
axes[0].imshow(img1)
axes[0].axis('off')
axes[1].imshow(img2)
axes[1].axis('off')
```

```python
#convert the images to grayscale
gray1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Display the images in the subplots
fig.suptitle('GRAYSCALE VERSIONS')
axes[0].imshow(gray1, cmap='gray')
axes[0].axis('off')
axes[1].imshow(gray2, cmap='gray')
axes[1].axis('off')
```

```python
# Ensure both images have the same dimensions
if gray1.shape != gray2.shape:
    gray2 = cv2.resize(gray2, (gray1.shape[1], gray1.shape[0]))

#check color difference between images
ssim_score = ssim(gray1, gray2)

print(f'SSIM VALUE = {ssim_score}\n')
if(ssim_score > 0.7):
    print('color difference test match PASSED')
else:
    print('color difference test match FAILED')
```
